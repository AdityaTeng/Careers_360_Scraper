```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://engineering.careers360.com/colleges/ranking/2019'
r = requests.get(url)
data = r.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
data_divs = soup.find_all('tr')
data_divs
data_divs = data_divs[1:101]

data_list = list()

for data_div in data_divs:
    data_dict = dict()
    data_dict['college'] = data_div.find('td', {'class':'colgName'}).text
    data_dict['nirf_rank'] = data_div.find('span', {'class':'nirf-rank'}).text
    data_dict['careers_rating'] = data_div.find('span', {'class':'rating greenFont'}).text
    data_dict['stream'] = 'engineering'
    data_list.append(data_dict)
data_list
```

```python
df = pd.DataFrame(data_list)
df
```

```python
df_1 = pd.pivot_table(df, index='careers_rating', values='stream', aggfunc='count', margins=True)
df_1
```

## getting data for all streams

```python
url_dict = {'engineering'      : 'https://engineering.careers360.com/colleges/ranking/2019',
            'mba_govt'         : 'https://bschool.careers360.com/colleges/ranking/2020/top-government-mba-colleges-in-India',
            'mba_private'      : 'https://bschool.careers360.com/colleges/ranking/2020/top-private-mba-colleges-in-India',
            'law_govt'         : 'https://law.careers360.com/colleges/ranking/2020/top-government-law-colleges-in-India',
            'law_private'      : 'https://law.careers360.com/colleges/ranking/2020/top-private-law-colleges-in-India',
            'medicine_govt'    : 'https://medicine.careers360.com/colleges/ranking/top-government-medical-colleges-in-India',
            'medicine_private' : 'https://medicine.careers360.com/colleges/ranking/top-private-medical-colleges-in-India'
           }
```

```python
data_list = list()
for stream in url_dict:
    url = url_dict[stream]
    print()
    print(stream)
    for page in range(1, 10):
        print(page, end=" ")
        params = {'page' : page}
        response = requests.get(url, params=params)
        data = response.text
        soup = BeautifulSoup(data, 'html.parser')
        data_divs = soup.find_all('tr')
        data_divs = data_divs[1:101]
        for data_div in data_divs:
            data_dict = dict()
            try:
                data_dict['college'] = data_div.find('td', {'class':'colgName'}).text
            except:
                data_dict['college'] = "-"
            try:
                data_dict['nirf_rank'] = data_div.find('span', {'class':'nirf-rank'}).text
            except:
                data_dict['nirf_rank'] = "-"
            try: 
                data_dict['careers_rating'] = data_div.find('span', {'class':'rating greenFont'}).text
            except:
                data_dict['careers_rating'] = "-"
            data_dict['stream'] = stream
            data_list.append(data_dict)
data_list
```

```python
df = pd.DataFrame(data_list)
```
